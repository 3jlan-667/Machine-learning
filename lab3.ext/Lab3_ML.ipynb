```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

# Laoding dataset 
df = pd.read_csv("sales.csv")
```

```python
# Displaying first 5 rows
df.head()
```

```python
# Check all set for nan valuse 
df.isna()
```

```python
# Count of nan values
print(df.isna().sum())
```

```python
# Check for duplicatetion
df.duplicated().sum()
```

```python
# Count of rows and columns 
print("Shape (rows, columns): ", df.shape,"\n")

print("number of rows: ", df.shape[0])
print("number of columns: ", df.shape[1])
```

```python
# data types of columns
df.dtypes
```

```python
# Statistical summary
df.describe(include='all')
```

```python
# Gender distrbtion 

gender_counts = df['gender'].value_counts()
plt.figure(figsize=(6,4))
plt.bar(gender_counts.index, gender_counts.values, color=['skyblue','pink'])
plt.title("Distribution of Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python
# Bar chart of Branch sales for each branch

branch_sales = df.groupby('branch')['total_price'].sum()
plt.figure(figsize=(6,4))
plt.bar(branch_sales.index, branch_sales.values)
plt.title("Total Sales by Branch")
plt.xlabel("Branch")
plt.ylabel("Total Sales")
plt.show()
```

```python
# Bar chart of Quantity Sold by Product Category
cat_quantity = df.groupby('product_category')['quantity'].sum()
plt.figure(figsize=(8,4))
plt.bar(cat_quantity.index, cat_quantity.values, color='green')
plt.title("Quantity Sold by Product Category")
plt.xlabel("Product Category")
plt.ylabel("Quantity Sold")
plt.show()
```

```python
# Bar chart of Distribution of Unit Price
plt.figure(figsize=(6,4))
plt.hist(df['unit_price'], bins=20, color='purple', edgecolor='black')
plt.title("Distribution of Unit Price")
plt.xlabel("Unit Price")
plt.ylabel("Frequency")
plt.show()
```

```python
# Scatter plot of Quantity vs Total Price
plt.figure(figsize=(6,4))
plt.scatter(df['quantity'], df['total_price'], alpha=0.6, color='red')
plt.title("Quantity vs Total Price")
plt.xlabel("Quantity")
plt.ylabel("Total Price")
plt.show()
```

```python
# Correlation Heatmap
sns.heatmap(df[['unit_price','quantity','tax','total_price','reward_points']].corr(),
            annot=True, cmap='coolwarm')
```
